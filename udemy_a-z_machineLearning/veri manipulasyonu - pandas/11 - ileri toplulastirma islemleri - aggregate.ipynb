## ileri toplulastirma islemleri - aggregate, filter, transform, apply


```python
import numpy as np
import pandas as pd

df = pd.DataFrame(
    {
        "gruplar": ["A", "B", "C", "A", "B", "C"],
        "degisken1": [10, 23, 33, 22, 11, 99],
        "degisken2": [100, 253, 333, 262, 111, 969],
    },
    columns=["gruplar", "degisken1", "degisken2"],
)
df
```

#### ========================================================

#### aggregate


```python
df.groupby("gruplar").mean()
```

```python
df.groupby("gruplar").aggregate([min, np.median, max])

# [] içinde bir liste gönderdik ve bu listeyi bütün değişkenlere uygulayacak.
```

```python
df.groupby('gruplar').aggregate({'degisken1': min,
                                 'degisken2': max})

# sözlük yapısının içinde degisken1 icin min, degisken2 icin max degerini getirmesini istedik
```
