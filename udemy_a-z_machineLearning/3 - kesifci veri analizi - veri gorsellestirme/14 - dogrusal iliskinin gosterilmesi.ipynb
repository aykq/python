## dogrusal iliskinin gosterilmesi


```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")
df = tips.copy()
df.head()

# scatterplot'taki sacilim ile gosterilen bilgiyi dogru araciligiyla temsil etmek icin kullandik
```

```python
sns.lmplot(x='total_bill', y='tip', data=df)

# lmplot icindeki lm = linear model

# dogrunun etrafindaki acik renkli alanlar sapmalari ifade ediyor
```

```python
sns.lmplot(x='total_bill', y='tip', hue='smoker', data=df)
```

```python
sns.lmplot(x='total_bill', y='tip', hue='smoker', col='time', data=df)
```

```python
sns.lmplot(x='total_bill', y='tip', hue='smoker',
           col='time', row='sex', data=df)
```
