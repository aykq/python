## scatterplot matrisi


```python
import seaborn as sns

iris = sns.load_dataset("iris")
df = iris.copy()
df.head()

# iris veri seti = cicek turleri ve ozellikleri
```

```python
df.dtypes

# 4 sayisal degisken
# 1 tane kategorik degisken
```

```python
df.shape

# 150 gozlem (satÄ±r), 5 degisken (sutun)
```

```python
sns.pairplot(df)

# satir ve sutun baz alinarak olusturulmus olan grafigin yapisal bir formu yoksa,
# o iki degisken arasinda bir iliski olmadigi anlamina gelir

# noktalar birbirinden farkli yerlerde kumeleniyorsa, caprazlama ile gozlemleyebilecek oldugumuz alt gruplar vardÄ±r
```

```python
sns.pairplot(df, hue='species')

# bir ustte yazilan farkli farkli gruplanmalarin ture gore renklendirildigi grafikler asagidaki gibidir
```

```python
sns.pairplot(df, hue='species', markers=['o', 's', 'D'])
```

```python
sns.pairplot(df, kind='reg')
```

```python
sns.pairplot(df, kind='reg', hue='species')
```
