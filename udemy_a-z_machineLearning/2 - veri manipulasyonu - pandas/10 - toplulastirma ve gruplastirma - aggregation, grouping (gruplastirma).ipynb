## toplulastirma ve gruplastirma - aggregation, grouping (gruplastirma)

### gruplama islemleri


```python
import pandas as pd
df = pd.DataFrame({'gruplar': ['A', 'B', 'C', 'A', 'B', 'C'],
                   'veri': [10, 11, 52, 23, 43, 55]}, columns=['gruplar', 'veri'])
df
```

```python
df.groupby("gruplar")
```

```python
df.groupby("gruplar").mean()
```

```python
df.groupby("gruplar").sum()
```

```python
import seaborn as sns

df = sns.load_dataset("planets")
df.head()
```

```python
df.groupby("method")
```

```python
df.groupby("method")["orbital_period"]
```

```python
df.groupby("method")["orbital_period"].mean()

# method kategorisinden unique olan değerleri filtreleyip, orbital period değişkenine göre ortalamasını aldı.
```

```python
df.groupby("method")["mass"].mean()
```

```python
df.groupby("method")["mass"].describe()
```

```python
df.groupby("method")["mass"].describe().T
```
