## sutun grafik caprazlamalar


```python
import seaborn as sns
from pandas.api.types import CategoricalDtype
diamonds = sns.load_dataset('diamonds')
df = diamonds.copy()
cut_kategoriler = ['Fair', 'Good', 'Very Good', 'Premium', 'Ideal']
df.cut = df.cut.astype(CategoricalDtype(
    categories=cut_kategoriler, ordered=True))
df.head()
```

### **cut kategorik degiskeni ile price sayisal degerini caprazlayarak, bir arada kullanarak gorsellestirme**


```python
sns.catplot(x='cut', y='price', data=df)

# catplot = kategorik degiskenleri gorsellestirmek icin kullanilan bir fonksiyon
# catplot = categoric plot
```

```python
sns.barplot(x='cut', y='price', hue='color', data=df)

# price eksenindeki degerler, 2 kategorik degisken 1 arada bulundugunda,
# 2 groupby islemi sonrasinda prica degiskeninin ortalamasi

# her sutunun tepesindeki cizgiler ise standart sapmayi (standart hata) temsil ediyor
```

```python
df.groupby(["cut", "color"])["price"].mean()

# ustteki grafikte price ekseninde cikan degerler buradan geliyor
```
