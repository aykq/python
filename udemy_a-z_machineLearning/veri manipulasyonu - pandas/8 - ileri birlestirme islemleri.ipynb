## ileri birlestirme islemleri


```python
import pandas as pd
```

#### ========================================================

#### birebir birlestirme


```python
df1 = pd.DataFrame(
    {
        "calisanlar": ["Ali", "Veli", "Ayse", "Fatma"],
        "grup": ["Muhasebe", "Muhendislik", "Muhendislik", "IK"],
    }
)
df1
```

```python
df2 = pd.DataFrame(
    {
        "calisanlar": ["Ayse", "Ali", "Veli", "Fatma"],
        "ilk_giris": [2010, 2009, 2014, 2019],
    }
)
df2
```

```python
pd.merge(df1, df2)
```

```python
pd.merge(df1, df2, on="calisanlar")

# verdiğimiz "on=" parametresi ile bu dataframe'leri "calisanlar" altında birleştirmesini söylüyoruz.
```

#### ========================================================

#### many to one - coktan teke birlestirme


```python
df3 = pd.merge(df1, df2)
df3
```

```python
df4 = pd.DataFrame(
    {
        "grup": ["Muhasebe", "Muhendislik", "IK"],
        "mudur": ["Caner", "Mustafa", "Berkcan"],
    }
)
df4
```

```python
pd.merge(df3, df4)
```

#### ========================================================

#### many to many - coktan coka birlestirme


```python
df5 = pd.DataFrame(
    {
        "grup": ["Muhasebe", "Muhasebe", "Muhendislik", "Muhendislik", "IK", "IK"],
        "yetenekler": ["matematik", "excel", "kodlama", "linux", "excel", "yonetim"],
    }
)
df5
```

```python
df1
```

```python
pd.merge(df1, df5)
```
