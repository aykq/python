## cizgi grafik - line plot


```python
import seaborn as sns

fmri = sns.load_dataset("fmri")
df = fmri.copy()
df.head()

# zamana bagli degiskenlerin oldugu veri setlerinde ve nesnelerin interneti gibi senaryolarda makinelerin
# urettigi verileri gorsellestirmek icin kullanbilabilir

# subject = verilerin toplandigi kisiler
# timepoint = zaman noktalari
# event = verinin toplanmasiyla ilgili birbirinden farkli olaylar
# (region) = sinyalin toplandigi bolge
# signal = gelen sinyal
```

```python
df.shape
```

```python
df["timepoint"].describe()
```

```python
df["signal"].describe()
```

```python
df.groupby("timepoint")["signal"].count()

# subjectlerin unique olup olmadigina bakiyoruz

# her zaman noktasinda esit sayida sinyal toplanmis
```

```python
df.groupby("signal").count()

# 1064 tane gozlem yani 1064 tane tekil signal degeri ciktigi icin, evet bu signaller unique diyebiliriz
```

```python
df.groupby("timepoint")["signal"].describe()
```

```python
sns.lineplot(x="timepoint", y="signal", hue="event", data=df)

# eklediÄŸimiz hue argÃ¼manÄ± ile birlikte peak yapan noktalarÄ± nelerin etkilediÄŸini gÃ¶rmÃ¼ÅŸ olduk
```

```python
sns.lineplot(data=df, x="timepoint", y="signal", hue="event", style="event")
```

```python
sns.lineplot(
    data=df,
    x="timepoint",
    y="signal",
    hue="event",
    style="event",
    markers=True,
    dashes=False,
)

# ortalama deÄŸerleri daha kolay gÃ¶rebilmek iÃ§in nokta halinde gÃ¶sterdik
# dashes = false, ortalamayÄ± gÃ¶steren Ã§izgileri kesikli olarak deÄŸil de dÃ¼z Ã§izgi olarak gÃ¶stermesi iÃ§in
```
