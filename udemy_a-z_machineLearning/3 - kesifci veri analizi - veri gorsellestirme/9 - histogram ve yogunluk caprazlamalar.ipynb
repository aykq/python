## histogram ve yogunluk caprazlamalar


```python
import seaborn as sns

diamonds = sns.load_dataset("diamonds")
df = diamonds.copy()
df.head()
```

```python
sns.kdeplot(df.price, fill=True)
```

```python
(sns
 .FacetGrid(df,
            hue='cut',
            height=5,
            xlim=(0, 10000))
 .map(sns.kdeplot, 'price', fill=True)
 .add_legend()
 )

# normalde 'ideal' olan sınıfın daha pahalı olanları oluşturduğu düşünülebilir fakat grafiğe bakıldığında
# ideal sınıf ürünlerinin daha çok 0-3000 arasında olduğu gözlemlenebilir.
```

```python
sns.catplot(data=df, x='cut', y='price', hue='color', kind='point')
```
