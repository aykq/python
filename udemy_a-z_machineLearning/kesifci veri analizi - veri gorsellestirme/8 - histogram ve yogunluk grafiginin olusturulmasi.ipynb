## histogram ve yogunluk grafiginin olusturulmasi


```python
import seaborn as sns

diamonds = sns.load_dataset("diamonds")
df = diamonds.copy()
df.head()
```

```python
# sns.distplot(df.price, kde=False);

# normalde egitimde ustteki gibi distplot kullaniliyor fakat seaborn yeni surumde distplot kaldirilacak
# distplot yerine displot kullanilmasi gerekiyor

sns.displot(df.price, kde=False)

# histogramin mantigi, elimizdeki sayisal degiskenin degerlerini belirli araliklara boler ve belirli araliklarda
# ilgili degerle ilgili gozlem(sutun) ve frekanslari yansitir.
```

```python
sns.displot(df.price, bins=10, kde=False)

# bins = olusturulacak histogramin kac sutunu olacagini belirler
# bins = 10 oldugu icin x ekseninde 10 sutun var
```

```python
sns.displot(df.price, bins=100, kde=False)

# yine ayni sekilde bins = 100 oldugu icin x ekseninde 100 tane sutun var
```

```python
sns.displot(df.price, kind='kde')
# sns.displot dokumaninda kullanimi bu sekilde gosterilmis

sns.displot(df.price, kde=True)
# eger istenirse histogram ile beraber de cizdirilebilir ancak y eksenindeki olcekler bozulabilir

# kde = True argumaniyla olasilik yogunluk grafigi eklendi
```
